## Capturing NAN value with new Feature

```python
import numpy as np
import pandas as pd
df=pd.read_csv("train.csv",usecols=["Survived","Age","Fare"])
```

```python
df.head()
```

```python
df["Age_new"] = np.where(df["Age"].isna(),1,0)
```

```python
df.head()
```

```python
df.Age.mean()
```

```python
df['Age'].fillna(df.Age.mean(),inplace=True)
```

```python
df.drop(["Age_new"],axis=1)
```
